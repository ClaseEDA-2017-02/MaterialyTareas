Como declara y usar una clase

```python
class stack:
    """Ejemplo de pila"""
    size=0
    datos=[]
    def __init__(self):
        self.datos = []
        self.size=0
    def push(self,n):
        self.datos=[n]+self.datos
        self.size+=1
    def pop(self):
        if self.size >0:
            res=self.datos[0]
            self.datos=self.datos[1:]
        return res
    def toString(self):
        s=""
        for i in self.datos:
            s+=i+" "
        return s
    
```

```python
miPila=stack()
```

```python
miPila.push('3')
miPila.push('2')
miPila.push('1')
```

```python
print miPila.toString()
```

```python
miPila.pop()
```

Lectura y procesamiento rapido de un archivo

```python
f=open('datosEj.txt') # poner la ruta y nombre del archivo
```

```python
datos=f.read()
# lee todo a un string. veando print datos
```

```python
Separados=datos.lower().strip('.,').split() # el argumento de split por default es esoacio. Me regresa una lista
```

Contar ocurrencias de cada palabra. Hacer un diccionario

```python
d={}
```

```python
for i in Separados:
    if i in d.keys():
        d[i]=d[i]+1
    else:
        d[i]=0
```

```python
sorted(d,key=lambda x: d[x], reverse=True)
```
